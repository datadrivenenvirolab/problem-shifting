# Merge city plans to target dataset

```python
from pathlib import Path
import pandas as pd

targets_path = Path("data_raw/climactor_targets_combined.csv")
plans_path = Path("data_raw/ClimActor_6_plans.csv")
output_path = Path("data_inter/climactor_targets_with_plans.csv")

# Load data
targets = pd.read_csv(targets_path)
plans = pd.read_csv(plans_path)
```

```python
# Plans dataset clean up and merge to targets
plan = (
    plans.rename(columns={"datasource_id": "plan_datasource"})
    .drop(columns=["plan_boundary_desc", "plan_url", "plan_comments", "plan_name"])
)

clean_plan = plan.loc[
    plan["plan_adopted_year"].notna() & plan["plan_type"].notna()
].copy()

target_plan = (
    targets.merge(clean_plan, on="climactor_id", how="left")
    .drop_duplicates()
)

print(f"Target plan rows: {len(target_plan)}")
target_plan.head() # 5379

target_plan.to_csv(output_path, index=False)
```
